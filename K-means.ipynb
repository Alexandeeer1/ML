```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from google.colab import drive
```

```python
drive.mount('/content/drive')
```

```python
Carpeta_Del_Dataset = '/content/drive/MyDrive/ML-Proyecto-Dataset'
Contenido_Del_Dataset = os.listdir(Carpeta_Del_Dataset)
for archivo in Contenido_Del_Dataset:
    print(archivo)
```

```python
imagenes = []

for i, subcarpeta in enumerate(Contenido_Del_Dataset):
    ruta_subcarpeta = os.path.join(Carpeta_Del_Dataset, subcarpeta)
    archivos_imagenes = os.listdir(ruta_subcarpeta)
    for archivo_imagen in archivos_imagenes:
        ruta_imagen = os.path.join(ruta_subcarpeta, archivo_imagen)
        imagen = cv2.imread(ruta_imagen)
        imagen_gris = cv2.cvtColor(imagen, cv2.COLOR_BGR2GRAY)
        imagenes.append(imagen_gris.flatten())
```

```python
matriz_caracteristicas = np.array(imagenes)
```

```python
kmeans = KMeans(n_clusters=24)
kmeans.fit(matriz_caracteristicas)
```

```python
labels = kmeans.labels_
centroides = kmeans.cluster_centers_
```

```python
for i, subcarpeta in enumerate(Contenido_Del_Dataset):
    ruta_subcarpeta = os.path.join(Carpeta_Del_Dataset, subcarpeta)
    archivos_imagenes = os.listdir(ruta_subcarpeta)

    for j, archivo_imagen in enumerate(archivos_imagenes):
        ruta_imagen = os.path.join(ruta_subcarpeta, archivo_imagen)
        imagen = cv2.imread(ruta_imagen)
        imagen = cv2.cvtColor(imagen, cv2.COLOR_BGR2RGB)
        indice = i*len(archivos_imagenes) + j
        if indice < len(labels):
            plt.subplot(len(Contenido_Del_Dataset), len(archivos_imagenes), indice + 1)
            plt.imshow(imagen)
            plt.axis('off')
            plt.title(f'Cluster {labels[indice] + 1}')
            plt.show()
```
